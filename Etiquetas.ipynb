```python
import pandas as pd
import os
import numpy as np
import scipy.stats as sts
import shutil
from os import listdir
```

```python
breed_dicc = {}
for breed in listdir('Dogs/all_data/'):
    breed_dicc[breed] = len(listdir('Dogs/all_data/'+breed))
```

```python
import operator
sorted_breed_dicc = sorted(breed_dicc.items(), key=operator.itemgetter(1))
```

```python
etiquetas= []
```

```python
for sorted_dicc in sorted_breed_dicc:
    raza= sorted_dicc[0] 
    etiquetas.append(raza)
```

```python
#invierte las razas en lista segun mayor número de ejemplos de entrenamiento hasta la de menor número
etiquetas=etiquetas[::-1] 
etiquetas
```

```python
dir_top= "Dogs/Ktop/"
dir_actual= "Dogs/all_data/"
```

```python
breeds_completo= etiquetas

def det_ids(breeds_completo, clases):
    razas_top=breeds_completo[0:clases] #razas_top contiene el nombre de las kclases con más ejemplos   
    #print (razas_top)
    for reg in listdir(dir_actual):        
        if reg in razas_top:
            cont=1
            for imagen in listdir(dir_actual+reg):
                try:
                    os.mkdir(dir_top+"all_train_"+str(clases)+'/')
                    os.mkdir(dir_top+"all_val_"+str(clases)+'/')
                    #os.mkdir(dir_top+"all_train_"+str(clases)+'/'+reg+'/')
                except Exception as e:
                    pass
                try:
                    #os.mkdir(dir_top+"ejemplos_train_"+str(clases)+'/')
                    os.mkdir(dir_top+"all_train_"+str(clases)+'/'+reg+'/')
                    os.mkdir(dir_top+"all_val_"+str(clases)+'/'+reg+'/')
                except Exception as e:
                    pass
                if cont<=int(len(listdir(dir_actual+reg))*0.8):
                    shutil.copy(dir_actual+reg+'/'+imagen, dir_top+"all_train_"+str(clases)+'/'+reg+'/'+imagen)
                else:
                    shutil.copy(dir_actual+reg+'/'+imagen, dir_top+"all_val_"+str(clases)+'/'+reg+'/'+imagen)
                    
                cont+=1
        
    return
```

```python
det_ids(breeds_completo, 5)
det_ids(breeds_completo, 10)
det_ids(breeds_completo, 15)
det_ids(breeds_completo, 20)
det_ids(breeds_completo, 25)
```
