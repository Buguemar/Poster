```python
import pandas as pd
import os
import numpy as np
import scipy.stats as sts
import shutil
from os import listdir

labels=pd.read_csv("Datasets/labels.csv")
labels.head()
```

```python
breeds= sts.itemfreq(labels['breed'])
breeds=sorted(breeds, key=lambda x: -x[1])
dir_top= "Dog Breed/train_top/"
dir_actual= "Dog Breed/train/"
```

```python
breeds[0:15]
```

```python
def det_ids(breeds_completo, clases):
    breeds_sub=breeds_completo[0:clases]
    #os.mkdir(dir_top)
    razas_top= []
    for reg in breeds_sub:
        razas_top.append(reg[0])
        
    for reg in listdir(dir_actual):        
        if reg in razas_top:
            cont=1
            for imagen in listdir(dir_actual+reg):
                try:
                    os.mkdir(dir_top+"ejemplos_train_"+str(clases)+'/')
                    os.mkdir(dir_top+"ejemplos_val_"+str(clases)+'/')
                   #os.mkdir(dir_top+"ejemplos_train_"+str(clases)+'/'+reg+'/')
                    print ("listo car")
                except Exception as e:
                        #print (e)
                        pass
                try:
                    #os.mkdir(dir_top+"ejemplos_train_"+str(clases)+'/')
                    os.mkdir(dir_top+"ejemplos_train_"+str(clases)+'/'+reg+'/')
                    os.mkdir(dir_top+"ejemplos_val_"+str(clases)+'/'+reg+'/')
                    print ("listo raza")
                except Exception as e:
                        #print (e)
                        pass
                if cont<=int(len(listdir(dir_actual+reg))*0.8):
                    shutil.copy(dir_actual+reg+'/'+imagen, dir_top+"ejemplos_train_"+str(clases)+'/'+reg+'/'+imagen)
                else:
                    print ("ESDEfv")
                    shutil.copy(dir_actual+reg+'/'+imagen, dir_top+"ejemplos_val_"+str(clases)+'/'+reg+'/'+imagen)
                    
                cont+=1
        
    return
```

#### Crea los directorios de entrenamiento segun las K clases de mayor nÃºmero de eje. training

```python
det_ids(breeds, 5)
det_ids(breeds, 10)
det_ids(breeds, 15)
det_ids(breeds, 20)
det_ids(breeds, 25)
```
